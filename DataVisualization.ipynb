### Scatter plots

```python
# load the new dataframe
uploaded = files.upload()
```

```python
df_insurance = pd.read_csv("insurance.csv") 

df_insurance.head(5)
```

Podemos ver que bmi y charges estan positivamente correlacionados.
Para verificar la fuerza de esta relacin podemos dibujar la linea que mejor ajusta a los datos

```python
sns.scatterplot(x=df_insurance["bmi"], y= df_insurance["charges"])
```

```python
sns.regplot(x=df_insurance['bmi'], y=df_insurance['charges'])
```

Podemos seguir analizando los datos y ver si hay alguna caracteristica que pueda separar aun mejor los datos.

Por ejm veamos como la caracteristica "smoker" influye en el aumento de "charges", para ello podemos plotear los puntos de dos colores, distinguiendo los fumadores de los no fumadores.

```python
sns.scatterplot(x=df_insurance['bmi'], y=df_insurance['charges'], hue=df_insurance['smoker'])
```

Para hacer mas evidente esto, podemos agregar dos regresiones lineales.

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=df_insurance)
```
